```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('dataset.csv')


print(df.info())
print(df.describe())


df.plot(x='Origin Display Name', y='Daily Mean Travel Time (Seconds)', kind='line')
plt.xlabel('Origin Display Name')
plt.ylabel('Daily Mean Travel Time (Seconds)')
plt.title('Travel time over differenet Locations')
plt.show()


df.plot(x='Origin Movement ID', y='Destination Movement ID', kind='scatter')
plt.xlabel('Origin Movement ID')
plt.ylabel('Origin Display Name')
plt.title('Origin ID v Origin Display name')
plt.show()
```
